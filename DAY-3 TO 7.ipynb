**LINE CHAET**

```python
import matplotlib.pyplot as plt
months=["jan","feb","mar","apr"]
sales=[12000,15000,18000,17000]
plt.plot(months,sales)
plt.title("monthly sales")
plt.xlabel("months")
plt.ylabel("Sales")
plt.show()
```

**PIE CHART**

```python
import matplotlib.pyplot as plt
students=["jan","feb","mar","apr"]
marks=[12,15,18,17]
plt.bar(students,marks)
plt.title("student marks")
plt.xlabel("students")
plt.ylabel("marks")
plt.show()
```

**BAR CHART**

```python
import matplotlib.pyplot as plt
expenses=[4000,3000,9000]
categories=["food","travel","shoping"]
plt.pie(expenses,
        labels=categories,
        autopct='%1.1f%%')
plt.show()
```

**line and bar chart**

```python
# Install required libraries (optional, usually pre-installed in Colab)
!pip install matplotlib pandas

# Import libraries

import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [23, 45, 56, 78]
}

df = pd.DataFrame(data)

# --- Line Chart ---
plt.figure(figsize=(6,4))
plt.plot(df['Category'], df['Values'], marker='o', color='blue')
plt.title('Line Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.grid(True)
plt.show()

# --- Bar Chart ---
plt.figure(figsize=(6,4))
plt.bar(df['Category'], df['Values'], color='green')
plt.title('Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()

# --- Pie Chart ---
plt.figure(figsize=(6,6))
plt.pie(df['Values'], labels=df['Category'], autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.show()
```
